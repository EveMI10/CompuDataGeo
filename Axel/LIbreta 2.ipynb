**I. JOSEPH FOURIER**    

J. Fourier nació en Auxerre, Fracia, en 1768. Después de una expedición por Egipto, Napoleón Bonaparte lo nombró *Barón* en 1808, y en 1809 terminó su *"Description d'Egypt"*, una obra que glorifica la arquitectura del Antiguo Egipto. Después de la derrota de Napoleón Bonaparte, en 1815 durante la *Batalla de Waterloo*, Fourier obtuvo puestos, como Director de la Oficina de Estadíticas para el Deartamento del Sena (1816) y Secretario Permanente de la Academia de Ciencias (*Académie Française*), en 1827. Murió en 1830, como resultado de complicaciones de una enfermedad atrapada en  Egipto (Gray, 2015). 
<br />
<br />
Sus trabajos son conocidos por la *descomposición de funciones periódicas en series trigonométricas* (**Series de Fourier**), método por el cual consiguió'resolver la *ecuación del calor*. La **Transformada de Fourier** recibe su nombre en su honor. Además, se le reconoce por ser el primero en dar una explicación científica sobre el *Efecto invernadero*.

**II. SERIES DE FOURIER**   

Las *Series de Fourier* surgen de la tarea práctica de representar una función periódica $f(t)$ dada, en términos de funciones *seno* y *coseno*. Dichas series son trigonométricas, cuyos coeficientes se determinan a partir de las *Fórmula de Euler* (Alcaraz-Candela, 2017). 
<br />
<br />
Si $f$ es una función real de periodo $2\pi$, tal que existen las integrales para $a_{0}$, $a_{n}$ y $b_{n}$, entonces se tiene la serie: 
<br />
$$\sum_{n=-\infty}^{\infty} c_{n}e^{inx} = a_{0} + \sum_{n=1}^{\infty} (a_{n}\cos{nx}+b_{n}\sin{nx})$$
<br />
$$a_{0}=\frac{1}{2\pi}\int_{-\pi}^{\pi}S(x)dx$$
<br />
$$a_{n}=\frac{1}{\pi}\int_{-\pi}^{\pi}S(x)\cos{nx}dx$$
<br />
$$b_{n}=\frac{1}{\pi}\int_{-\pi}^{\pi}S(x)\sin{nx}dx$$
<br />
<br />
Si una función es real $f$ y $2\pi$-periódica, es monónota a trozos y acocotada en [$-\pi$,$-\pi$]. Es decir, la *serie de Fourier* de $f$ converge en todos los puntos de los *números reales* (Teleco, 2005).
<br />
<br />
**III. TRANSFORMADA DE FOURIER**   

Para una función no periódica $P \rightarrow \infty$:
<br />
$$F(\omega) = \int_{-\infty}^{\infty} f(t) e^{-i\omega t}dt$$
<br />
$$F(\omega) = \frac{1}{2\pi} \int_{-\infty}^{\infty} F(\omega) e^{i\omega t}d\omega$$
<br />
<br />
La primer integral que obtiene $F(\omega)$ se denomina *Transformada de Fourier* de $f(t)$, y la segunda, *Transformada inversa de Fourier*. El cuadrado $f^{2}(t)$ nos da una idea de cómo la energía contenida en la onda se distribuye en el tiempo, mientras que $F^{2}(\omega)$ nos conduce a cómo la energía se distribuye en el espectro de frecuencias. 
<br />
<br />
**IV. TEOREMA DE NYQUIST**  

El *Teorema de muestreo* o *Teorema de Nyquist-Shannon* propone que una señal $x(t)$ con un espectro limitado a la frecuencia $f_{B} (|f| < f_{B})$ puede ser muestreada sin pérdida de información si la frecuencia de muestreo $f_{S}$ supera la cantidad $2f_{B}$. Es decir, $f_{S} \geq 2f_{B}$.
<br />
<br />
Si no se mestrea, como mínimo, a esa frecuencia, tiene lugar el fenómeno denominado *"aliasing"*. Por otra parte, si se muestrea de acuerdo al teorema, existe un proceso de reconstrución que garantiza la reproducción exacta de la señal continua $x(t)$, a partir de sus nuestras $x[n]$ (Ortega-García, 2003). 
<br />
<br />
**IV. CONVOLUCIÓN**  

La *convolución* de dos funciones $f(x)$ y $g(x)$ es la función resultante calculando, para cada valor de $x$, el valor ponderado de $f(x)$ con los valores de $g(x)$. En el caso de que el área encerrada por la curva de $g(x)$ fuera igual a $1$, entonces estaríamos calculando, para $x$, una media ponderada. Matemáticamente, la expresión para esta operación es:
<br />
$$f(x) * g(x) = h(x) = \int_{-\infty}^{\infty}f(z)g(x - z)dz$$

<br />
<br />
**IV. CORRELACIÓN**  

La *correlación* de dos funciones reales es una operación similar a la *convolución*, salvo que no se gira alrededor del origen los valores de una de las funciones. La expresión matemática es:
<br />
$$f(x)◦g(x) = h(x) = \int_{-\infty}^{\infty} f(x)g(x + z)dz$$

<br />
<br />
**V. FILTROS**  

Un filtro hace mención a *todo procesado que altera la naturaleza de una señal, de una forma u otra*. Un filtro es un proceso computacional o algorítmico, mediante el cual una señal (secuencia de muestras) es transformada en una segunda secuencia de muestras. Los filtros más comunes son los filtros pasa bajo (*Low Pass*), pasa alto (*High Pass*) y pasa banda (*Band Pass*). 
<ul>
<li>Los filtros *Low Pass* dejan pasar las frecuencias que están por debajo de una determinada frecuencia;</li>
<li>Los filtros *High Pass* dejan pasar las frecuencias que están por encima de una determinada frecuencia; </li>
<li>Los filtros *Band Pass* dejan pasar las frecuencias que están situadas en una determinada banda de frecuencias, es decir, entre dos determinadas frecuencias; y</li>
<li>Los *promedios móviles* son promedios calculador a partir de subgrupos artificiales de observaciones consecutivas. En los análisis de series de tiempo se pueden utilizar los promedios móviles para suavizar los datos y reducir las fluctuaciones aleatorias en una serie de tiempo.</li>
</ul>

Los filtros *Low Pass* y *High Pass* están definidos por su **frecuencia de corte**, que es la frecuencia a la cual la amplitud de la señal se reduce a $1/\sqrt{2}$ de su valor máximo. Por su parte, el filtro *Band Pass* está definido por su **frecuencia central** y su **ancho de banda**, que se resume a la diferencia entre las frecuencias de corte inferior y superior (Gómez-Gutiérrez, 2009). 
<br />
<br />
**VI. CAUSALIDAD Y NO CAUSALIDAD** 
<br />
Un sistema es *causal* si su salida para cualquier valor de la variable independiente depende únicamente del valor de la entrada, correspondiente a dicho valor. Todos los sistemas sin memoria son causales (Sánchez, 2005). 
<center>**Ejemplo de un sistema causal**</center>
$$f(t) = x(t - t_{0})$$
<br />
<br />
Por el contrario, un sistema *no causal* muestra una salida anticipada a la señal de entrada. La salida depende de los valores que $x$ tendría al momento $t+1$. 
<center>**Ejemplo de un sistema no causal**</center>
$$f(t) = x(t + t_{0})$$

<br />
<br />
**VII. ANÁLISIS TIEMPO-FRECUENCIA** 
<br />
En el procesamiento de señales, el *análisis tiempo-frecuencia* es un conjunto de técnicas para la caracterización y manipulación de señales cuyas estadísticas varían en el tiempo, como las señales transitorias.
<br />
<br />
Es una generalización y perfeccionamiento del análisis de Fourier, en el que las frecuencias son constantes en el tiempo. Debido a que muchas señales de interés tienen frecuencias cambiantes, el análisis tiempo-frecuencia tiene un alcance amplio.

```julia
using Plots
using DSP
using LaTeXStrings
```

```julia
#Frecuencia característica a los 5 Hz

nsec = 2
sps = 30
fify = sps/2
freq = 5
time = collect(linspace(0,nsec,(nsec*sps)))

##############

y = sin.(2*pi*freq*time)
#y = sin.(time1,freq)

##############

ft = abs.(fft(y,1));
ftp = ft[1:convert(Int64,(length(ft)/2))]
fvec = collect(linspace(0,fify,length(ftp)))

p1 = plot(fvec,ftp,color="steelblue")

pyplot()
plot(p1, label="Curva",
    grid=1, gridcolor="grey", gridlinestyle=:dash, gridalpha=1,
    legendfont=Plots.font("Palatino",12), xlabel=L"Frecuencia ($Hz$)", ylabel=L"Amplitud ($m$)", ylims=(0,29),
    title="Relación Frecuencia-Amplitud",
    box=:true)
```

```julia
#Frecuencias características a los 10 y 15 Hz

nsec = 2
sps1 = 40
fify1 = sps1/2
freq1 = 10
freq2 = 15
time1 = collect(linspace(0,nsec,(nsec*sps1)))

##############

y1 = sin.(2*pi*freq1*time1)
y2 = sin.(2*pi*freq2*time1)
yt = y1+y2

p2 = plot(time1,yt, color="steelblue")
#y = sin.(time1,freq)

plot(p2, label="Curva",
    grid=1, gridcolor="grey", gridlinestyle=:dash, gridalpha=1,
    legendfont=Plots.font("Palatino",12), xlabel=L"Tiempo ($s$)", ylabel=L"Amplitud ($m$)", ylims=(-1.99,1.99),
    title="Relación Tiempo-Amplitud",
    box=:true)
```

```julia
#Frecuencias características a los 10 y 15 Hz, con ruido

g = rand(length(yt))
y_noise = yt + g

##############

ft1 = abs.(fft(y_noise,1));
ftp1 = ft1[1:convert(Int64,(length(ft1)/2))]
fvec1 = collect(linspace(0,fify1,length(ftp1)))

##############

p3 = plot(fvec1,ftp1, color="steelblue")

plot(p3, label="Curva", 
    grid=1, gridcolor="grey", gridlinestyle=:dash, gridalpha=1,
    legendfont=Plots.font("Palatino",12), xlabel=L"Frecuencia ($Hz$)", ylabel=L"Amplitud ($m$)", ylims=(0,37),
    title="Relación Frecuencia-Amplitud",
    box=:true)
```

```julia
#Filtro LowPass
reponsetype = Lowpass(5; fs = 40)
design = Butterworth(8)
sisfil = filt(digitalfilter(reponsetype, design), MBHE);

p16 = plot(sisfil, title=L"Filtro $Low Pass$",color="steelblue", label="MBHE",xlabel=L"N")

plot(p16, 
    grid=1, gridcolor="grey", gridlinestyle=:dash, gridalpha=1,
    legendfont=Plots.font("Palatino",12), 
    box=:true)
```

```julia
#FFT al LP

ft17 = abs.(fft(sisfil,1))
ftp17 = ft17[2:convert(Int64,(length(ft17)/2))]
fvec17 = collect(linspace(0,20,length(ftp17)))
#p7 = plot(fvec3,ftp3, ylims=(0,310000))
p17 = plot(fvec17,ftp17,title=L"$FFT$ al filtro $Low Pass$",color="steelblue", label="FFT MBHE",xlabel=L"Frecuencia ($Hz$)")

plot(p17, 
    grid=1, gridcolor="grey", gridlinestyle=:dash, gridalpha=1,
    legendfont=Plots.font("Palatino",12), 
    box=:true)
```

```julia
#Filto HP
reponsetypeh = Highpass(5; fs = 40)
designh = Butterworth(8)
sisfilh = filt(digitalfilter(reponsetypeh, designh), MBHE);

p17 = plot(sisfilh, ylabel="Magnitud",xlabel=L"N", label="MBHE",title=L"Filtro $High Pass$",color="steelblue")

plot(p17, 
    grid=1, gridcolor="grey", gridlinestyle=:dash, gridalpha=1,
    legendfont=Plots.font("Palatino",12), 
    box=:true)
```

```julia
#Filtro BP
reponsetypeb = Bandpass(10,15; fs = 40)
designb = Butterworth(8)
sisfilb = filt(digitalfilter(reponsetypeb, designb), MBHE)
p18 = plot(sisfilb,ylabel="Magnitud",xlabel=L"N", label="MBHE",title=L"Filtro $Band Pass$",color="steelblue")

plot(p18, 
    grid=1, gridcolor="grey", gridlinestyle=:dash, gridalpha=1,
    legendfont=Plots.font("Palatino",12), 
    box=:true)
```
