<h3>Paleomagnetismo</h3>
<div style="text-align: justify">
El Paleomagnetismo es el estudio del campo magnético Terrestre registrado por las rocas en el momento de su formación o durante procesos geológicos relevantes acaecidos con posterioridad (p.e. metamorfismo) (Pueyo, 2008).
<br />
<br />
Se basa en tres suposiciones:
<ul>  
<li>Dipolo geocéntrico axial: Produce el campo magnético terrestre a través de procesos magneto-hidrodinámicos en el núcleo externo líquido; también contribuye la componente no dipolar.</li>    

<li>Registro del campo magnético primario: Los minerales
ferromagnéticos contenidos en las rocas son capaces de registrar el campo magnético terrestre del pasado.</li>  

<li>Registro invariante en el tiempo: La magnetización adquirida por una roca en el momento de su formación puede permanecer invariable en el tiempo</li>
</ul>


<h3>Componentes</h3>

Un vector en paleomagnetismo se puede descomponer en 3:
<img style="float: right;" src="compo.png" width="250">
<ul>  
<li>Inclinación($I$): ángulo entre el vector de intensidad ($H$) y la componente horizontal, va de $-90°-90°$, es positivo hacia abajo.</li>  
<li>Declinación($D$): Ángulo entre el norte geográfico y la componente horizontal, va de $0°-360°$, es positivo en el sentido horario. </li>
<br />
<li>Magnitud($H$): El cual tiene una componente vertical ($H_v$) y una horizontal ($H_h$) que tienene componentes geográficas, por lo tanto tenemos:</li> 
    
$$H_V=HsenI$$
    
$$H_h=HcosI$$
    
  Donde $H_v$ se compone de:
    
$$H_N=H cos I cos D$$
$$H_E=H cos I sin D$$
    
   Por lo que la intensidad total del campo está dada por:
    
$$H=\sqrt{H_N^2 + H_E^2 + H_V^2 }$$
</ul>
</div>

<h3>Campo magnético y Variación secular</h3>
<div style="text-align: justify">
<img style="float: right;" src="dircamp.png" width="350">

El campo magnético terrestre se encuentra en constante cambio, el estudio de sus variaciones en el tiempo es una forma de tratar de explicar los cambios en el interior de la Tierra, que dan origen al campo geomagnético, por lo que es necesario el desarrollo de registros regionales y globales (Juárez, 2016).
<br />
<br />
Los cambios en las componentes del vector geomagnético (declinación ($D$), inclinación ($I$) e intensidad ($H$)) tienen periodos de $1-10^5$ años (Butler, 2004), a este fenómeno se le conoce como variación secular (VS). Su origen de acuerdo con Butler (2004) se puede deber a la periodicidad supersupuesta de los cambios no dipolares en periodos cortos y en periodos largos.
<br />
<br />
De acuerdo con Butler (2004), los patrones de variación secular son similares en las regiones subcontinentales y muy distitos de un continente a otro.
<br />
<br />
Juárez (2016) menciona que existen curvas regionales de escala subcontinental que se encuentran en lugares específicos y distribuidos globalmente, que se construyen estadísticamente a partir de las tres componentes del campo geomagnético ($D, I, H$), resultando en tres curvas separadas. Estas curvas representan promedios suavizados de datos obtenidos en distintos lugares, a partir de datos de observatorios, así como mediciones paleomagnéticas de materiales arqueológicos y geológicos.
<br />
<br />
Existen modelos geomagnéticos globales como el modelo SHA.DIF.14k, el cual se construye a partir de de datos arqueomagnéticos y volcánicos disponibles del Holoceno y del cual fueron extraidos los datos de este trabajo. **Valores estadísticos**

```julia
#Suma, promedio y desviación estándar
    
function pyd3(x)
    su=0
    
    for i=1:length(x)
        su=su + x[i]
        
    end
    prom=su/length(x)
    de=0
    for i=1:length(x)
        sud=(x[i]-prom).^2
        de=(de + sud)
        
        
    end
    
    de=sqrt(de/(length(x)-1))
    
    println("Suma=", su," ", "Promedio=",prom, " ", "DesEst= ", de) 
    
end
```

```julia
pyd3(Edad) #Edad
pyd3(Declin) #Declinación
pyd3(Inclin) #Inclinación
pyd3(Alpha95) #Alpha95
```
