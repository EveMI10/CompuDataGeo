<h3>Teorema de Nyquist/Teorema de muestreo</h3>
<div style="text-align: justify">
De acuerdo a Alvarado y Stern (2010), este teorema es considerado el más importante en la adquisición de señales, ya que establece una condicipon necesaria y suficiente para la reconstrucción, en el dominio temporal, de una señal adquirida, estableciendo que la frecuencia de muestreo debe ser al menos 2 veces mayor que la frecuencia más alta de la señal que se quiere reconstruir:
<br />
<br />
$$f_s ≥ 2 f_m$$
<br />
Además menciona que la muestra es el conjunto de datos adquiridos; el periodo de muestreo o la resolución temperoral ($\Delta t$) es el intervalo de timepo entre cada dato que se adquiere, se recomienda que sea pequeño para que no pasen inadvertidos eventos importantes; $N$ es el número de puntos que se quiere adquirir.
<br />
<br />
El tiempo total $T_t$ de la muestra es igual a $T_t=N \Delta t$, éste debe de ser suficiente para observar completamente el fenómeno. Al inverso del período de muestreo se la llama frecruencia de muestreo $f_s=1/\Delta t$.
<br />
<br />
La resolución en el espacio de las frecuencias, es decir, el intervalo de frecuencias más pequeño que se puede estudiar es: $\Delta f = f_s / N$.
<br />
Cuando no se muestrea con la frecuencia correcta pueden aparecer componentes no deseadas de la señal original; a este fenómeno se le llama **aliasing**, para evitarlo se recomienda hacer un sobre muestreo.

<h3>Transformada de Fourier</h3>

Consiste en la descomposición de las señales en sumatorias de componentes senoidales. Esto proporciona la descripción de una señal dada, en función de la frecuencia. Un objetivo importante de la descomposición señalada, es la descripción de la distribución de la energía o de la potencia de una señal dada (y de la respuesta), en términos de tales frecuencias.

La **Transformada Discreta de Fourier** (DFT), involucra un proceso de discretización, considera a una señal no periódica x(n) como si fuese periódica a efecto de algoritmos de cálculo. Escobar (2008) menciona que tiene un período igual a la logitud N de la señal usada para la transformación.

La DFT está definida por:

$$X(k)= \displaystyle\sum_{n=0}^{N-1} x(n) e^{-j(2\pi kn/N)}$$

Donde:
k=0, 1, 2, ..., N-1, corresponde al índice del espectro discreto.
n=0, 1,  , ..., N-1, es el índice del tiempo discreto.

La **Transformada rápida de Fourier** (FFT) se desarrolla a partir de la definición de la DFT, la FFT es una forma eficiente de calcular la DFT y su éxito se debe a la reducción del número de adiciones y multiplicaciones requeridas para su cálculo (Escobar, 2008).

La FFT está definida por:

$$DFT (A)[K]= \displaystyle\sum_{n=1}^{lenght(A)} e^{-i{2\pi(n-1)(k-1)\over lenght(A)}} A[n]$$



<h3>Filtros</h3>

De acuerdo a Palomera (2002), un **filtro** permite el paso de señales que se encuentran dentro de un rano de frecuencias, la señal que no se encuentre dentro de ese rango será atenuada, en el caso ideal tendrá un valor de cero. Al rango de frecuencias que deja pasar el filtro se le conoce como **banda de paso o ancho de banda**, a la frecuencia limite del rango se le denomina **frecuencia de corte**.

<ul>
<li>**Filtro pasa bajos**: Es el más sencillo y tiene una ganancia unitaria para las frecuencias comprendidas entre 0 y la frecuencia de corte, y un valor de 0 para cualquier otra frecuencia.</li>

<li>**Filtro pasa altos**: Permite el paso de señales con frecuencias mayores a la frecuncia de corte</li>

<li>**Filtro pasa bandas**: Posee dos frecuencias de corte, una inferior $(f-1)$ y otra superior $(f1)$, dando como resultado una frecuencia central y un ancho de banda comprendido entre estas dos frecuencias.</li>
</ul>

**Smoothing**: operación que intenta eliminar varaciones de corto plazo de una señal para revelar las tendencias a largo plazo.

Un smoothing común es el **promedio móvil**, el cual calcula la media de los valores previos de n para algún valor de n. 

El promdedio móvil proporciona una forma de "suavizar" las variaciones drásticas, removiendo los cambios más extremos, de forma que se pueda apreciar su tendencia. Esto se logra al “desplazar” los valores medios aritméticos en la serie de tiempo.

Saca el promedio cada x puntos y lo agrega a la variación para suavizarla.

Los resulados del promedio móvil se pueden ajustar a una recta para encontrar la tendencia lineal de los datos originales.

<h3>Convolución</h3>

De acuerdo a Escobar (2008), la operación convolución entre señales discretas está compuesta de suma de productos y se le puede considerar la operación básica o fundamental del PDS (Procesamiento Digital de Señales). La operación de convolucipon de manera simbólica es:

$$y[n]=x[n]*h[n]$$

$$x(n)=	 \displaystyle\sum x(n) \delta(n-k)$$

$x(n)$ representa la superposicipon de versiones de un conjunto de funciones elemnetales, o sea, los impulsos unitarios desplazados $\delta(n-k)$.

Desarrollo suma de convolución o superposición (discreto).

$$y(n)=	 \displaystyle\sum_{i=0}^∞ x(i)h(n-i)$$

$$y(n)=x(0)h(n)+x(1)h(n-1)+x(2)h(n-2)+...+x(i)h(n-i)$$

La contraparte continua de la suma de convolución es la integral de convolucipon o superposición:

$$\displaystyle\int_{-∞}^{+∞} x(\tau) h(t-\tau)\, d\tau$$

Es conmutativa,asociativa y distributiva.

<h3>Correlación</h3>

Operación matemática que se parece a la convolución.

El objetivo de la correlación entre dos señales es medir que tanto se parecen entre sí para extraer alguna información, calcular retardos entre señales, verificar y calcular la periodicidad de una señal, calcular la energía de una señal.

Si x(n) y y(n) son dos señales de energá finita entonces la correlción entre $x(n)$ y $y(n)$ se define como:


$$r_{xy}(l)=\displaystyle\int_{n=-∞}^{∞} x(n) y(n-l); l=0, ±1, ±2...$$

$$r_{xy}(l)=\displaystyle\int_{n=-∞}^{∞} x(n+l) y(n); l=0, ±1, ±2...$$

Donde:

l=Índice de corrimiento en el tiempo
xy=Indica que x está fijo y y se desplza en l unidades.

De forma similar se puede escribir $r_{yx}(l)$:

$$r_{yx}(l)=\displaystyle\int_{n=-∞}^{∞} y(n) x(n-l); l=0, ±1, ±2... $$

$$r_{yx}(l)=\displaystyle\int_{n=-∞}^{∞} y(n+l) x(n); l=0, ±1, ±2...$$

Comparando las ecuaciones se observa que $r_{xy}(l)=r_{yx}(-l)$, esto significa que $r_{yx}(l)$ es el doble de $r_{xy}(l)=r_{yx}(-l)$.

Entonces la correlación de $r_{xy}(l)$ se puede escribir como la convolución:

$$r_{xy}(l)=x(l)*y(-l)$$


Desarrollo de la correlación:

$$r_{xy}(l)= \displaystyle\sum_{n=0}^∞ x(n)y(n-l)$$

$$r_{xy}(l)= \displaystyle\sum_{n=0}^∞ x(n+l)y(n); l=0, ±1, ±2...$$

<h3>Sistema causal y no causal</h3>

Escobar (2008) menciona que un sistema es causal si éste no reacciona hasta que le es aplicado un estímulo, es decir, en un sistema no se puede esperar el efecto de la transformación sin la existencia previa de una causa.

Se expresa como:

$$y(n)=f{x(n), x(n-1), x(n-2), x(n-3), ...}$$

Si las condiciones anteriores no se satisfacen el sistema es no causal.

**Nyquist**

```julia
nsec=2#Número de segundos
sps1=40 #Muestreo
fNy1=sps1/2;
frec1=10 #Frecuencia1
frec2=15 #Frecuencia2
tiempo1=collect(linspace(0,nsec,(nsec*sps1)));
```

```julia
#y=Asen(2pi*f*t)
y1=sin.(2pi*frec1*tiempo1)
y2=sin.(2pi*frec2*tiempo1)
yt=(y1+y2);
```

```julia
plot(tiempo1,yt, ylab="Amplitud", xlab="Tiempo", title="Amplitud vs Tiempo", box=:true, label="Datos", color="chartreuse3",xticks=(0:0.25:2), ylims=(-2,2.05), gridalpha=.4, grid=1.5, grid=:true, gridcolor="black")
#Gráfica de las 2 frecuencias
```

```julia
#Transformada de Fourier

fu=abs.(fft(yt,1));
fup=fu[1:convert(Int64,(length(fu)/2))] #Para quitar espejo
furvec1=collect(linspace(0,fNy1,length(fup)/2));
```

```julia
plot(furvec1,fup,xlab="Frecuencia(Hz)", ylab="Amplitud", title="Amplitud vs Frecuencia", box=:true, label="Datos con FFT", color="cyan3", xlims=(0,20.5), xticks=(0:2.5:20), ylims=(0,38), yticks=(0:5:35), gridalpha=.4, grid=1.5, grid=:true, gridcolor="black")
```

```julia
plot(furvec2,fup2,xlab="Frecuencia(Hz)", ylab="Amplitud", title="Amplitud vs Frecuencia", box=:true, label="Datos con ruido y FFT", color="deeppink4", xlims=(0,20.2), xticks=(0:2.5:20), ylims=(-0,20), yticks=(0:2.5:20), gridalpha=.4, grid=1.5, grid=:true, gridcolor="black")
```

**Aplicando Fourier para obtener frecuencias y agregando filtro de promedio móvil**

```julia
ft1 = abs.(fft(ABHE,1));
ftp1 = ft1[2:convert(Int64,(length(ft1)/2))] #Para que no tome en cuenta el elemento 1
fvec1 = collect(linspace(0,20,length(ftp1)))

p1F = plot(fvec1,ftp1, ylab="Amplitud", title="Amplitud vs Frecuencia", box=:true, label="E-O", color="deepskyblue1", gridalpha=.4, grid=1.5, grid=:true, gridcolor="black", xticks=(0:5:20,[]))

#Filtro
v1=ones(400)*(1/400) #Ventana de tiempo
c1=filt(v1,1,ftp1); #Convolución de datos y ventana de tiempo
p1F=plot!(fvec1,c1, line=1, label="Media móvil")

plot(p1F);
```
